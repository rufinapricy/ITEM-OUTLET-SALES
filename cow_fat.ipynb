```python
from sklearn.preprocessing import LabelEncoder
```

```python
to_be_encoded=['Item_Identifier','Item_Fat_Content','Item_Type','Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type']
```

```python
label_encode=LabelEncoder()
```

```python
for i in to_be_encoded:
    if i not in df.columns:
        print(f"Column '{i}' not found in DataFrame.")
        continue
    df[i]=label_encode.fit_transform(df[i])
```

```python
y=df['Item_Outlet_Sales']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.25)
```

```python
lr.fit(X_train,y_train)
```

```python
Y_predict=lr.predict(X_test)
```

```python
from sklearn import metrics
r2_train = metrics.r2_score(y_test
                            , Y_predict)
```
